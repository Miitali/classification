## defining KNN from scratch

```python
from collections import Counter
k = 7
def predict(x):
    # Compute Euclidean distances between x and all examples in the training set
    distances = [np.linalg.norm(x - x_train) for x_train in X_train.values]

    # Sort by distance and return indices of the first k neighbors
    k_indices = np.argsort(distances)[:k]

    # Extract the labels of the k nearest neighbor training samples
    k_nearest_labels = [np.array(y_train)[i] for i in k_indices]
    #print(k_nearest_labels)
    # Return the most common class label
    most_common = Counter(k_nearest_labels).most_common(1)
    return most_common[0][0]
```

```python
predicted = [predict(x) for x in X_test.values]
print(predicted)
```

```python
cm = confusion_matrix(y_test, predicted)
print(cm)
print(classification_report(y_test, predicted))
```

## Task1: use user-defined k-NN classifier for mice and breast-cancer data

```python
mice_data = fetch_openml(name='miceprotein', version=4)
print(mice_data.data.shape)
print(mice_data.target.shape)
print(np.unique(mice_data.target))
print(mice_data.DESCR)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from collections import Counter

# Define the k-NN prediction function
def predict_knn(X_train, y_train, X_test, k):
    # Broadcasting creates an array of shape (n_train, n_test, n_features)
    distances = np.linalg.norm(X_train[:, np.newaxis] - X_test, axis=2)

    # For each test sample, get the indices of the k nearest training samples.
    # np.argsort returns indices along the training samples axis.
    k_indices = np.argsort(distances, axis=0)[:k]

    # Retrieve the labels for the k nearest neighbors.
    k_nearest_labels = np.array([y_train[idx] for idx in k_indices])

    # For each test sample, determine the most common label among the k nearest neighbors.
    predictions = [Counter(k_nearest).most_common(1)[0][0] for k_nearest in k_nearest_labels.T]
    return predictions

##########################
# MICE PROTEIN DATASET
##########################

# Load the Mice Protein dataset
mice_data = fetch_openml(name='miceprotein', version=4)

# Convert the data and target to numpy arrays
X_mice = mice_data.data.to_numpy()
y_mice = mice_data.target.to_numpy()

# Split data into training and testing sets
X_train_mice, X_test_mice, y_train_mice, y_test_mice = train_test_split(
    X_mice, y_mice, test_size=0.2, random_state=42
)

# Define k for k-NN
k = 7


predicted_mice = predict_knn(X_train_mice, y_train_mice, X_test_mice, k)

# Evaluate the Mice Protein model
cm_mice = confusion_matrix(y_test_mice, predicted_mice)
print("Mice Protein Confusion Matrix:")
print(cm_mice)
print(classification_report(y_test_mice, predicted_mice))

##########################
# BREAST CANCER DATASET
##########################

# Load the Breast Cancer dataset
breast_cancer_data = fetch_openml(name='breast-cancer', version=1)

# Convert every column to numeric by factorizing its values.
# This will replace each column with numeric codes.
breast_cancer_data.data = breast_cancer_data.data.apply(
    lambda col: pd.factorize(col.astype(str))[0]
)

# Also factorize the target labels
breast_cancer_data.target = pd.factorize(breast_cancer_data.target.astype(str))[0]

# Convert the DataFrame to numpy arrays for consistency
X_bc = breast_cancer_data.data.to_numpy()
y_bc = breast_cancer_data.target  # already numeric

# Split data into training and testing sets
X_train_bc, X_test_bc, y_train_bc, y_test_bc = train_test_split(
    X_bc, y_bc, test_size=0.2, random_state=42
)

# Predict on the Breast Cancer dataset
predicted_bc = predict_knn(X_train_bc, y_train_bc, X_test_bc, k)

# Evaluate the Breast Cancer model
cm_bc = confusion_matrix(y_test_bc, predicted_bc)
print("\nBreast Cancer Confusion Matrix:")
print(cm_bc)
print(classification_report(y_test_bc, predicted_bc))
```
